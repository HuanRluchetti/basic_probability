Exerc√≠cio 1: Dado viciado (mais chances de sair 6)

```python
import random
import matplotlib.pyplot as plt
import numpy as np

pesos = [1, 1, 1, 1, 1, 2]

def jogar_dados():
    """Simula o lan√ßamento de dois dados e retorna a soma."""
    dado1 = random.randint(1, 6)
    dado2 = np.random.choice([1, 2, 3, 4, 5, 6], p=np.array(pesos) / sum(pesos))
    return dado1 + dado2

def calcular_probabilidade(n=100000):
    """Calcula a probabilidade da soma dos dados ser 7 ap√≥s N lan√ßamentos."""
    contagem_7 = sum(1 for _ in range(n) if jogar_dados() == 7)
    return contagem_7 / n

def gerar_grafico(n=100000):
    """Gera um gr√°fico de distribui√ß√£o das somas dos dados."""
    resultados = [jogar_dados() for _ in range(n)]
    plt.hist(resultados, bins=range(2, 14), edgecolor='black', alpha=0.7, density=True)
    plt.xticks(range(2, 13))
    plt.xlabel('Soma dos dados')
    plt.ylabel('Frequ√™ncia relativa')
    plt.title(f'Distribui√ß√£o das somas dos dados ({n} lan√ßamentos)')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

if __name__ == "__main__":
    n_simulacoes = 100000
    probabilidade = calcular_probabilidade(n_simulacoes)
    print(f"Probabilidade estimada de obter soma 7: {probabilidade:.4f}")
    gerar_grafico(n_simulacoes)
```

Em uma lanchonete, 60% dos clientes pedem refrigerante. Qual a probabilidade de dois clientes seguidos pedirem refrigerante?

```python
import random
import matplotlib.pyplot as plt
import numpy as np

chance = [40, 60]

def jogar_dados():
    cliente1 = np.random.choice([1, 2], p=np.array(chance) / sum(chance))
    cliente2 = np.random.choice([1, 2], p=np.array(chance) / sum(chance))
    return cliente2 + cliente2

def calcular_probabilidade(n=100000):
    contagem = 0
    last = True
    for i in range(n):
      if jogar_dados() == 4:
          if last:
            contagem += 1
          else:
            last = True
      else:
        last = False

    return contagem / n

def gerar_grafico(n=100000):
    resultados = [jogar_dados() for _ in range(n)]
    plt.hist(resultados, bins=range(1, 10), edgecolor='black', alpha=0.7, density=True)
    plt.xticks(range(1, 10))
    plt.xlabel('Soma dos dados')
    plt.ylabel('Frequ√™ncia relativa')
    plt.title(f'Distribui√ß√£o ({n} lan√ßamentos)')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

if __name__ == "__main__":
    n_simulacoes = 100000
    probabilidade = calcular_probabilidade(n_simulacoes)
    print(f"Chance de 2 clientes seguidos pedirem refri: {probabilidade:.4f}")
    gerar_grafico(n_simulacoes)
```

Em uma pesquisa, 80% dos entrevistados preferem produto A ao produto B. Qual a probabilidade de tr√™s entrevistados aleat√≥rios preferirem A

```python
import random
import matplotlib.pyplot as plt
import numpy as np

chance = [80, 20]

def jogar_dados():
    cliente1 = np.random.choice([1, 2], p=np.array(chance) / sum(chance))
    cliente2 = np.random.choice([1, 2], p=np.array(chance) / sum(chance))
    cliente3 = np.random.choice([1, 2], p=np.array(chance) / sum(chance))
    return cliente1 + cliente2 + cliente3

def calcular_probabilidade(n=100000):
    contagem = sum(1 for _ in range(n) if jogar_dados() == 3)
    return contagem / n

def gerar_grafico(n=100000):
    resultados = [jogar_dados() for _ in range(n)]
    plt.hist(resultados, bins=range(1, 10), edgecolor='black', alpha=0.7, density=True)
    plt.xticks(range(1, 10))
    plt.xlabel('Soma dos dados')
    plt.ylabel('Frequ√™ncia relativa')
    plt.title(f'Distribui√ß√£o ({n} lan√ßamentos)')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

if __name__ == "__main__":
    n_simulacoes = 100000
    probabilidade = calcular_probabilidade(n_simulacoes)
    print(f"A chance de 3 cliente preferirem o produto A: {probabilidade:.4f}")
    gerar_grafico(n_simulacoes)
```

Se uma l√¢mpada tem uma vida √∫til que segue uma distribui√ß√£o exponencial com m√©dia de 1000 horas, qual a probabilidade de durar mais de 1200 horas?

```python
import math

def calcular_probabilidade_exponencial(media, t):
    """ Calcula a probabilidade de uma l√¢mpada durar mais que t horas, dado que segue distribui√ß√£o exponencial. """
    lambd = 1 / media  # Taxa da distribui√ß√£o exponencial
    return math.exp(-lambd * t)

if __name__ == "__main__":
    media_vida_util = 1000  # M√©dia de 1000 horas
    tempo_desejado = 1200   # Queremos saber a probabilidade de durar mais que 1200 horas

    probabilidade = calcular_probabilidade_exponencial(media_vida_util, tempo_desejado)
    print(f"Probabilidade de a l√¢mpada durar mais de {tempo_desejado} horas: {probabilidade:.4f}")
```

Qual a probabilidade de, em um grupo de 23 pessoas, pelo menos duas terem anivers√°rio no mesmo dia?

```python
import math
import random

def probabilidade_aniversario(n_pessoas):
    """ Calcula a probabilidade de pelo menos duas pessoas no grupo terem o mesmo anivers√°rio. """
    prob_diferentes = 1.0
    dias_no_ano = 365

    for i in range(n_pessoas):
        prob_diferentes *= (dias_no_ano - i) / dias_no_ano

    return 1 - prob_diferentes

def simular_aniversarios(n_pessoas, n_simulacoes=100000):
    """ Simula grupos de n_pessoas e verifica a frequ√™ncia de pelo menos duas pessoas com o mesmo anivers√°rio. """
    contador = 0

    for _ in range(n_simulacoes):
        aniversarios = [random.randint(1, 365) for _ in range(n_pessoas)]
        if len(aniversarios) != len(set(aniversarios)):  # Se houver repeti√ß√£o
            contador += 1

    return contador / n_simulacoes

if __name__ == "__main__":
    n_pessoas = 23
    prob_teorica = probabilidade_aniversario(n_pessoas)
    prob_simulada = simular_aniversarios(n_pessoas)

    print(f"Probabilidade te√≥rica de pelo menos duas pessoas compartilharem anivers√°rio: {prob_teorica:.4f}")
    print(f"Probabilidade simulada com Monte Carlo: {prob_simulada:.4f}")
```
